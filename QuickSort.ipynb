```python
import matplotlib.pyplot as plt
import random
import time
from array import array

def partition(arr,si,ei):

    pivot=arr[ei]
    i=si-1

    for j in range(si,ei):
        if(arr[j]<=pivot):
            i+=1
            temp=arr[j]
            arr[j]=arr[i]
            arr[i]=temp

    i+=1
    temp=arr[i]
    arr[i]=pivot
    arr[ei]=temp

    return i


def quickSort(arr,si,ei):
    if(si>=ei):
        return
    pivIdx=partition(arr,si,ei)
    quickSort(arr,si,pivIdx-1)
    quickSort(arr,pivIdx+1,ei)

# Input sizes to test
input_sizes = list(range(0,200001,15000))  # Up to 200000  


execution_times = []


# Measure execution time for each input size
for size in input_sizes:
    arr=[]
    # Generate a random array of given size
    for _ in range(size):
        num=random.randint(0,size)
        arr.append(num)
        
    # Record the time taken to sort
    start_time = time.time()
    quickSort(arr,0,len(arr)-1)
    end_time = time.time()

    # Store time in milliseconds
    elapsed_time = (end_time - start_time) * 1000
    execution_times.append(elapsed_time)
    print(f"Size: {size}, Time: {elapsed_time:.4f} ms")

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot([0, 175000], [0, 150], color='red', linestyle='-', linewidth=2, label='Linear O(n)')
plt.plot(input_sizes, execution_times, marker='', linestyle='-', color='blue', label='Quick Sort O(nLog(n))')
plt.title("Quick Sort Time Complexity Analysis")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (ms)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
    

    
```
