```python
import matplotlib.pyplot as plt
import random
import time
from array import array

# Bubble Sort Algorithm
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

# Input sizes to test
input_sizes = list(range(70, 7001, 700))  # Up to 7000  

execution_times = []

# Measure execution time for each input size
for size in input_sizes:
    # Generate a random array of given size
    arr = [random.randint(0, size) for _ in range(size)]

    # Record the time taken to sort
    start_time = time.time()
    bubble_sort(arr)
    end_time = time.time()

    # Store time in milliseconds
    elapsed_time = (end_time - start_time) * 1000
    execution_times.append(elapsed_time)
    print(f"Size: {size}, Time: {elapsed_time:.4f} ms")

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot([0, 7000], [0, 500], color='red', linestyle='-', linewidth=2, label='Estimated O(n)')
#I have shifted the slope of line so that comparasion becomes easier 
#also changine the slope of line does not interfere with linear property of line
plt.plot(input_sizes, execution_times, marker='', linestyle='-', color='blue', label='Bubble Sort')
plt.title("Bubble Sort Time Complexity Analysis")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (ms)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
